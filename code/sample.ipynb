```python
from langchain_community.llms import LlamaCpp

# モデルをロード
llm = LlamaCpp(
    model_path="mistral-7b.Q4_K_M.gguf", # ←ここのモデルを変える
    n_ctx=4096,      # 文脈長（入力の最大トークン数）
    n_threads=4,     # CPUスレッド数
    temperature=0.3, # 出力のランダム性
    verbose=False
)

# プロンプト
prompt = f"""
次の文章を300文字以内の日本語で要約してください。

{text}
"""

# 推論実行
output = llm.invoke(prompt)

print("=== 要約結果 ===")
print(output)
```
